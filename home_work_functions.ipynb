### Домашнее задание к лекции "Функции"

```python
# перечень всех документов
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# перечень полок, на которых хранятся документы
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


def inp_com():  # Ввод команды
    """вводим команду в виде строки,
     возвращаем команду """
    return input('Введите команду: ')


def inp_numb_doc():  # Ввод номера документа
    """ вводим номер документа,
    возвращаем номер документа"""
    return input('Введите номер документа: ')


def prn_res():  # Печатаем  'Результат:' и переносим строку
    print()
    print('Результат:')
    print()


def owner_doc(n):  # Выдаёт имя владельца  по  номеру документа.
    """Выдаёт имя владельца  по  номеру документа."""
    prn_res()
    for i in documents:
        if i['number'] == n:
            return print('Владелец документа:', i['name'])
    return print('Документ не найден в базе')


def dir_doc(n):  # может по номеру документа узнать на какой полке он хранится
    """принимает номер документа,
    выдаёт на какой полке он хранится"""
    for i in directories:
        if n in directories[i]:
            return i


def prn_dir_doc(n):  # Проверяет наличие полки и выдаёт её.
    """Принимает номер полки.
    Проверяет наличие полки и выдаёт её"""
    prn_res()
    if n in directories:
        print('Документ хранится на полке:', n)
    else:
        print('Документ не найден в базе.')


def all_doc():  # позволяет увидеть полную информацию по всем документам
    """ позволяет увидеть полную информацию по всем документам"""
    for i in documents:
        print('№:', i['number']+',', 'тип:', i['type']+',', 'владелец:',
              i['name']+',', 'полка хранения:', dir_doc(i['number']))


def inp_dir():  # вводим и возвращаем номер полки
    """ вводим номер полки """
    n = input('Введите номер полки:')
    return n


def check_dir(n):  # проверяем существует ли с таким номером полка
    """ проверяем существует ли с таким номером полка,
    вернет True если полка существует, и вернет False если не существует номер полки """
    return n in directories


def dir_now():  # Текущий перечень полок.
    """ выдаёт текущий перечень полок в определенном формате печати"""
    print('Текущий перечень полок: ', end = (''))
    ret = 0
    for i in directories:
        print(i, end = (''))
        ret += 1
        if ret != len(directories):
            print(', ', end = '')
    print('.')


def add_dir(n):  # Добавляем полку
    """ Добавляем полку с номером n """
    prn_res()
    if check_dir(n):
        print('Такая полка уже существует. ', end = '')
        return
    else:
        directories[n] = []
        print('Полка добавлена. ', end = '')


def del_dir(n):  # Удаляем полку
    """ Удаляем полку с номером n ,
    предварительно проверяем , есть ли в ней что-то"""
    prn_res()
    if check_dir(n):  # проверяем наличие полки
        if len(directories[n]) != 0:  # проверяем наличие документов на  полке
            print('На полке есть документы, удалите их перед удалением полки. ', end = '')
            return
        else:
            del directories[n]
            print('Полка удалена. ', end = '')
            return
    else:
        print('Такой полки не существует. ', end = '')


def inp_doc():  # Вводим новый документ
    """Вводим номер документа, тип документа, владельца документа,
    и выдаём их"""
    return {'type': input('Введите тип документа:'), 'number': input('Введите номер документа:'), 'name': input('Введите владельца документа:')}


def add_doc():  # Добавляем документ на полку
    """ Добавляем документ на заданную полку,
    предварительно проверяем наличие полки"""
    new_d = inp_doc()
    dir_inp = inp_dir()
    prn_res()
    if check_dir(dir_inp):
        prom_d = directories[dir_inp]
        prom_d.append(new_d['number'])
        directories[dir_inp] = prom_d
        documents.append(new_d)
        print('Документ добавлен. ', end = '')
        return
    else:
        prn_tpns()
        print('Добавьте полку командой ads.')


def prn_vst():  # печатаем  'Текущий список документов:'
    print('Текущий список документов:')


def d_doc():  # удаляем документ, предварительно проверяем его существование
    """ удаляем документ по введеному номеру, предварительно проверяем его существование """
    n = inp_numb_doc()
    prn_res()
    for i in documents:
        if i['number'] == n:
            directories[dir_doc(i['number'])].remove(i['number'])
            documents.remove(i)
            return print('Документ удален.')
        else:
            f_k = 'Документ не найден в базе.'
    print(f_k)


def prn_tpns():  # Печатаем 'Такой полки не существует.'
    print('Такой полки не существует. ', end = '')


def m_doc():  # Перемещаем документ с одной полки на другую
    """ Перемещаем документ с одной полки на другую, предварительно проверяем его существование,
     и существование полки"""
    numb_doc = inp_numb_doc()
    dir_inp = inp_dir()
    prn_res()
    if check_dir(dir_inp):
        for i in documents:
            if i['number'] == numb_doc:
                directories[dir_doc(i['number'])].remove(i['number'])
                directories[dir_inp].append(i['number'])
                print('Документ перемещен.')
                prn_vst()
                all_doc()
                return
            else:
                f_k = 'Документ не найден в базе'
        print(f_k)
        prn_vst()
        all_doc()
        return
    prn_tpns()
    dir_now()


# Основная ветвь:
in_sym = ()
while in_sym != 'q':
    print()
    in_sym = inp_com()
    if in_sym == 'p':
        owner_doc(inp_numb_doc())
    elif in_sym == 's':
        prn_dir_doc(dir_doc(inp_numb_doc()))
    elif in_sym == 'l':
        prn_res()
        all_doc()
    elif in_sym == 'ads':
        add_dir(inp_dir())
        dir_now()
    elif in_sym == 'ds':
        del_dir(inp_dir())
        dir_now()
    elif in_sym == 'ad':
        add_doc()
        prn_vst()
        all_doc()
    elif in_sym == 'd':
        d_doc()
        prn_vst()
        all_doc()
    elif in_sym == 'm':
        m_doc()
print('Завершение работы.')
exit(1)
```
