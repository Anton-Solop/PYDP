###### 1. Задание 1 <a id='task1'></a>

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic

для источников paid и email из России - ставим ad

для источников paid и email не из России - ставим other

все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
def source_type(row):
    """ Принимает строку и возвращает значения по следующей логике :
    - если источник traffic_source равен yandex или google, то в source_type ставится organic
    - для источников paid и email из России - ставим ad
    - для источников paid и email не из России - ставим other
    - все остальные варианты берем из traffic_source без изменений
    """
    if row['traffic_source'] == 'yandex' or row['traffic_source'] == 'google':
        # если источник traffic_source равен yandex или google, то в source_type ставится organic
        return 'organic'
    elif row['traffic_source'] == 'paid' or row['traffic_source'] == 'email':
        if row['region'] == 'Russia':
            # для источников paid и email из России - ставим ad
            return 'ad'
        # для источников paid и email не из России - ставим other
        return 'other'
    # все остальные варианты берем из traffic_source без изменений
    return row['traffic_source']
```

```python
# Добавляем столбец  'source_type' с помощью функции def source_type (row):
log['source_type'] = log.apply(source_type, axis=1)
log.head(20)
```

```python
log.source_type.value_counts()
```

###### 3. Задание 3 <a id='task3'></a>

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
# создаём Series где ключ - userId , значение - количество строк с этим userId
quantity = ratings.userId.value_counts()
```

```python
# добавляем столбец  'numb_rat_hund'  с значением True если пользователь выставил более 100 оценок
ratings['numb_rat_hund'] = ratings.apply(
    (lambda row: quantity[row['userId']] > 100), axis=1)
# отсортировываем строки с 'numb_rat_hund'== True , считаем время жизни для каждой строки, усредняем эти значения по всем строкам и присваиваем res_
res_ = ratings.loc[ratings['numb_rat_hund'] == True].groupby('userId').agg({'timestamp': [
    lambda x: x.max()-x.min()]}).mean()
print(
    'Cреднее время жизни пользователей, которые выставили более 100 оценок:', res_[0])
```

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе Материалы для лекции «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»).

Необходимо сформировать две таблицы:

  -- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента

  -- аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# Применяем объединение 'outer' , чтобы по условиям в процессе объединения таблиц данные не терялись
# хотя в данных условиях можно использовать объединение 'left' соблюдая последовательность объединений

# Объединяем датафреймы  client_base и air в resalt_
resalt_ = client_base.merge(air, how='outer', on='client_id')
# Объединяем датафреймы  resalt_ и auto
resalt_ = resalt_.merge(auto, how='outer', on='client_id')
# Объединяем датафреймы  resalt_ и rzd
resalt_ = resalt_.merge(rzd, how='outer', on='client_id')
# Заменяем NaN на ноль
resalt_ = resalt_.fillna(0)
resalt_
```

```python
# Убираем столбец 'address'
resalt_non_adr = resalt_.drop(['address'], axis=1)
resalt_non_adr
```
